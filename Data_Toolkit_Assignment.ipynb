1. What is the difference between multithreading and multiprocessing ?

MultiThreading:

  - Running multiple threads (smaller units of a process) within the same process
  - Threads share the same memory space
  - Concurrency (good for I/O-bound tasks)
  - Limited by the GIL (Global Interpreter Lock) — only one thread runs Python bytecode at a time
  - Low overhead (lightweight)

MultiProcessing:

  - Running multiple processes, each with its own Python interpreter
  - Each process has its own memory space
  - True Parallelism (good for CPU-bound tasks)
  - Each process has its own GIL, so they can truly run in parallel
  - High overhead (starting new processes takes more resources)

2. What are the challenges associated with memory management in Python ?

   - Key Challenges in Memory Management:

      a. Reference Cycles (Cyclic References)
         - When two or more objects refer to each other, their reference counts never drop to zero.

      b. Memory Leaks
         - Some references never get deleted (due to lingering variables, closures, or caching).

      c. Fragmentation of Memory
         - Python allocates small objects in non-contiguous blocks, causing fragmentation.

      d. Global Interpreter Lock (GIL) Effects
         - Only one thread can execute Python bytecode at a time, limiting true parallel memory management.

   

3. Write a Python program that logs an error message to a log file when a
division by zero exception occurs ?

```python
import logging

# Configure logging
logging.basicConfig(
    filename="error_log.txt",      # Log file name
    level=logging.ERROR,           # Log only ERROR and above
    format="%(asctime)s - %(levelname)s - %(message)s"
)

def divide_numbers(a, b):
    try:
        result = a / b
        print(f"Result: {result}")
    except ZeroDivisionError as e:
        logging.error("Division by zero error occurred: %s", e)
        print("Error: Division by zero is not allowed!")

# Example usage
divide_numbers(10, 0)
```

4. Write a Python program that reads from one file and writes its content to
another file.


```python
# Program to read from one file and write its content to another file

# Specify source and destination file names
source_file = "source.txt"
destination_file = "destination.txt"

try:
    # Open the source file in read mode
    with open(source_file, "r") as src:
        # Read the entire content
        content = src.read()

    # Open the destination file in write mode
    with open(destination_file, "w") as dest:
        # Write the content to the destination file
        dest.write(content)

    print(f"Content successfully copied from '{source_file}' to '{destination_file}'.")

except FileNotFoundError:
    print(f"Error: The file '{source_file}' was not found.")

except IOError as e:
    print(f"I/O error occurred: {e}")
```

5. Write a program that handles both IndexError and KeyError using a
try-except block.

```python
# Program to handle IndexError and KeyError using try-except

def handle_errors():
    my_list = [10, 20, 30]
    my_dict = {"a": 1, "b": 2, "c": 3}

    try:
        # Intentionally cause IndexError
        print("Accessing list element at index 5:", my_list[5])

        # Intentionally cause KeyError
        print("Accessing dictionary key 'd':", my_dict["d"])

    except IndexError as ie:
        print("IndexError occurred:", ie)

    except KeyError as ke:
        print("KeyError occurred:", ke)

    finally:
        print("Execution completed (finally block executed).")

# Run the function
handle_errors()
```

6. What are the differences between NumPy arrays and Python lists ?

1. Numpy Arrays:

    - All elements must be of the same type
    - Much faster (stored as contiguous C arrays)
    - More memory-efficient
    - Supported elementwise (np.array([1,2,3]) * 2 → [2,4,6])
    - Huge library of math, linear algebra, stats, etc

2. Python Lists:

    - Can store elements of different types ([1, "two", 3.0])
    - Slower (elements are stored as Python objects)
    - Consumes more memory
    - Not directly supported ([1,2,3] * 2 → [1,2,3,1,2,3])
    - Limited built-in functions

7. Explain the difference between apply() and map() in Pandas ?

1. apply():

     - Works on Series and DataFrame
     - Row-wise, column-wise, or element-wise
     - Returns Series or DataFrame
     - Accepts Function only
     - Use for Complex operations on rows/columns

2. map():

     - Works on Series only
     - Element-wise operation
     - Returns Series
     - Accepts Function, dict, Series
     - Use for Simple transformations

8. Create a histogram using Seaborn to visualize a distribution.


```python
# ✅ Compare Two Distributions using Seaborn Histogram

import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Generate random data for two groups
np.random.seed(42)  # For reproducibility

male_scores = np.random.normal(loc=72, scale=8, size=250)    # Mean=72, SD=8
female_scores = np.random.normal(loc=68, scale=10, size=250) # Mean=68, SD=10

# Create a DataFrame
df = pd.DataFrame({
    'Scores': np.concatenate([male_scores, female_scores]),
    'Gender': ['Male'] * 250 + ['Female'] * 250
})

# Create histogram with KDE for both groups
sns.histplot(data=df, x='Scores', hue='Gender', kde=True, bins=20, palette='Set2', alpha=0.6)

# Add titles and labels
plt.title("Distribution of Student Scores by Gender", fontsize=14)
plt.xlabel("Score", fontsize=12)
plt.ylabel("Frequency", fontsize=12)

# Add legend and show plot
plt.legend(title="Gender")
plt.show()
```
